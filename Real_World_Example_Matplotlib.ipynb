```python
gas = pd.read_csv('gas_prices.csv')
gas.head()
```

```python
# Clean Empty Data
gas['Australia'].fillna(gas['Australia'].mean(), inplace=True)
```

```python
gas.isnull().sum()
```

```python
plt.close('all')
```

```python
plt.close()

plt.figure(figsize=(8,5))

# Plotting
plt.title('Gas Price over Time (in USD)', fontdict={'fontweight':'bold'})

# Plotting the lines and adding labels for the legend

for country in gas:
    if(country != "Year"):
        plt.plot(gas.Year, gas[country], marker=".", label=country)

# plt.plot(gas.Year, gas.USA, 'b.-', label='United States')  # USA prices
# plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')  # Canada prices
# plt.plot(gas['Year'], gas['South Korea'], 'g.-', label='South Korea')  # South Korea prices
# plt.plot(gas.Year, gas.UK, 'y.-', label='United Kingdom')  # UK prices

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollar')

plt.legend()

plt.show()

plt.savefig('Gas_Price_Graph.png', dpi=300)

plt.close('all')
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
# HistoGram !
```

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='g')

plt.xticks(bins)

plt.xlabel("Skill Label")
plt.ylabel("Number of Player")
plt.title("Distribution of Players skills in FIFA (2018)", fontdict={'fontweight':'bold'})

plt.show()
```

```python
######## Pie Chart #########
```

```python
# left_count = fifa[fifa['Preferred Foot'] == "Left"].count()[0]
left_count = fifa[fifa['Preferred Foot'] == "Left"].shape[0]
left_count

# Right_count = fifa[fifa['Preferred Foot'] == "Right"].count()[0]
Right_count = fifa[fifa['Preferred Foot'] == "Right"].shape[0]
Right_count

label = ['Left', 'Right']
color = ['#fefae0','#aabb34']

plt.pie([left_count,Right_count], labels=label, colors=color, autopct='%.2f %%')
plt.title("Foot Prefrences of FIFA Player (2018)", fontdict={'fontweight':'bold'})

plt.show()
```

```python
fifa.Weight
```

```python
fifa['Weight'] = fifa['Weight'].where(
    fifa['Weight'].isna(),
    fifa['Weight'].str.replace('lbs', '').astype(float)
)
```

```python
fifa.Weight
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
heavy_medium = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]
```

```python
weight = [light,light_medium,medium,heavy_medium,heavy]
label = ['Under 125','125-150','150-175','150-175','Over 200']
color = ['#9C89B8','#F0A6CA','#EFC3E6','#F0E6EF','#B8BEDD']

plt.figure(figsize=(8,8))
explode = (.4,.1,0,0,.4)

plt.pie(weight, labels=label, colors=color, autopct='% 2f %%', explode=explode)

plt.title("Weight Distribution of FIFA Player (in lbs)", fontdict={'fontweight':'bold'})

plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

plt.boxplot([barcelona, madrid, revs], labels=labels)

plt.title("Professional Soccer Team Comparison")
plt.ylabel("FIFA Overall Rating (2018)")

plt.show()
```
