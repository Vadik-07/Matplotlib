```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Basics Graph
```

```python
x = [0,1,2,3,4,5,6]
y = [0,2,4,6,8,10,12]

# Resize Your Graph (dpi specify pixel per inch. when saving probably should use 300 if possible)
plt.figure(figsize=(7,5), dpi=100)

# Line 1

# Keyword Argument Notation
# plt.plot(x,y, label='2x', color='Red', linewidth=4, marker='.', markersize=15, markeredgecolor="Blue", linestyle="--")

# Shorthand Notation
# plt.plot(x,y, 'b^--', label='2x')
plt.plot(x,y, 'r.--', label='2x')

# Line 2

# Select Interval we want to plot point as
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2, label="x^2") 

# Add a Title (Specify font Parameter with fontdict)    
plt.title("My First Graph!", fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})

# X and Y Lables
plt.xlabel("X Axis", fontdict={'fontname' : 'Comic Sans MS'})
plt.ylabel("Y Axis", fontdict={'fontname' : 'Comic Sans MS'})

# X, Y axis Tickmarks (scale of your Graph)
plt.xticks([0,1,2,3,4,5,6])
# plt.yticks([0,2,4,6,8,10,12,14])

# Add a Legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has High Resolution)
plt.savefig('Myfirstgraph.png', dpi=300)

# Show Plot
plt.show()
```

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

pattern = ['/','O','*']
for bar in bars:
    bar.set_hatch(pattern.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize=(5,3))

plt.show()
```
